# Result Analysis 

We are going to visualize the SQL result to better analysis.

```python
# importing necessary libraries
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('bmh')

%matplotlib inline
```

## 1. Top Paying Jobs

```python
df_1 = pd.read_json('1_top_paying_jobs.json')
df_1
```

```python
plt.figure(figsize=(10,7))
plot = sns.barplot(df_1, x='average_salary', y='job_title_short')
plot.bar_label(plot.containers[0], fontsize=10)
plot.set(xlabel = 'Average Salary', ylabel = '', title = 'Top Paying Jobs')
plt.xlim(0,200000)
plt.savefig('../assets/1_top_paying_jobs.png', bbox_inches = 'tight')
```
